```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(13,10))

X_top = [1.3, 2.8]
Y_top = [0.0097, 0.0104]
Er_top = [0.01147736 - 0.0097, 0.01085534 - 0.0104]

X_vote = [1.7, 3.2]
Y_vote = [0.0130, 0.0137]
Er_vote = [0.01660213 - 0.0130, 0.01433662 - 0.0137]

color_yyahn = ['#e33831', '#f7ad30', '#8dc53e', '#cab6d9', '#08b07f', 
            '#f2a794', '#ed643a', '#871918', '#a1d9e8', '#b985ba', 
            '#d5c726', '#508dc4', '#2b2064', '#7d7d7d', '#7852A9', 
            '#1db58f', '#faccbd', '#dc2227', '#609f5c', '#cd6c74', 
            '#f293d1', '#f4a720', '#a08f27']

c1 = '#7f6baf'
c2 = '#f0832f'

plt.errorbar(X_top , Y_top ,yerr=Er_top, fmt='s',ms = 12, lw = 4, capsize = 8, c = c1, label = 'top answer (binary) ~ task experience', capthick=4, alpha = 1)
plt.errorbar(X_vote , Y_vote ,yerr=Er_vote, fmt='s',ms = 12, lw = 4, capsize = 8, c = c2, label = '#votes ~ task experience', capthick=4, alpha = 1)

plt.plot(X_top , Y_top ,'s',ms = 8, c = c1)
plt.plot(X_vote , Y_vote ,'s',ms = 8, c = c2)

x_label_name = ''
y_label_name = 'effect of experience'

plt.xlabel(x_label_name,fontsize=30)
plt.ylabel(y_label_name,fontsize=30)
plt.title('# votes or top answer (binary) ~ task experience',fontsize=40, fontdict={'weight': 'bold'})
# ####
plt.tick_params(labelsize=30)
#plt.legend()
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
ax=plt.gca();#
ax.spines['bottom'].set_linewidth(4);###
ax.spines['left'].set_linewidth(4);####
ax.spines['right'].set_linewidth(4);###
ax.spines['top'].set_linewidth(4);####
#ax.set_xscale('log')
#ax.set_yscale('log')
plt.ylim([0,0.017])
plt.xlim([1,3.5])
plt.xticks([1.5,3],['year fixed effects','year and task fixed effect'])
plt.yticks([0.0,0.005,0.010,0.015],[0.0,0.005,0.010,0.015])
#  X  
plt.grid(axis='y',linestyle='--', color = '#7d7d7d',lw = 3, alpha = 0.23)
#plt.xticks([],[])

font1 = {
        'weight' : 'bold',
        'size'   : 30,
}
legend = plt.legend(prop=font1, frameon = False)

#plt.savefig('output/figure_3c_top_vote_regression.svg', format = 'svg', transparent = True, bbox_inches = 'tight')
plt.show()
```
